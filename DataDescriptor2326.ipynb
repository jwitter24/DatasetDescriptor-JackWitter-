```python
import pandas as pd

df=pd.read_csv("my_dataset.csv")

df.head()



```

```python
df.shape
```

```python
df_clean=df.dropna(subset=['FULLADDRESS', 'Date'])
```

```python
df_clean.shape
```

```python
df_clean['FULLADDRESS'] = df_clean['FULLADDRESS'].str.upper()
df_clean = df_clean[df_clean['FULLADDRESS'].str.contains(r"\bNW\b", na=False)]

df_clean.shape
df_clean['FULLADDRESS'].head(10)
```

```python
streetsfocus = [
    "LOUGHBORO","MACARTHUR","ARIZONA","CANAL","DALECARLIA","POTOMAC","NORTON", 
    "SHERIER","NEWARK","MANNING","MACOMB","CAROLINA","CATHEDRAL","GALENA","WEAVER",
    "KLINGLE","HAWTHORNE","LOWELL","MILLWOOD","51ST","GLENBROOK","MAUD","PALISADE",
    "WATSON","PARTRIDGE","UNIVERSITY","CHAIN BRIDGE","NEBRASKA","GARFIELD","HURST",
    "ESKRIDGE","BATTERY","GLADE","FULTON","DANA","DORSETT","EDMUNDS"]
```

```python
pattern = "|".join(streetsfocus)

df_clean = df_clean[df_clean['FULLADDRESS'].str.contains(pattern, na=False)]
```
